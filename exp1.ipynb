```python
import pandas as pd
import numpy as np
import seaborn as sns                       
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("sample_data/california_housing_test.csv")
df.head()
```

```python
df.tail()
```

```python
df.dtypes
```

```python
df.rename(columns= {"total_rooms" : "rooms"})
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
df.count()
```

```python
df = df.drop_duplicates()
df.count()
```

```python
print(df.isnull().sum)
```

```python
df = df.dropna()
df.count()
```

```python
print(df.isnull().sum)
```

```python
sns.boxplot(x = df["median_house_value"])
```

```python
df.housing_median_age	.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
fig, ax = plt.subplots(figsize=(20,12))
ax.scatter(df['housing_median_age'], df['median_house_value'])
ax.set_xlabel('housing_median_age')
ax.set_ylabel('Price')
plt.show()
```
